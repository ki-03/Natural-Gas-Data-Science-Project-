```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df_gas.set_index('Dates')['Prices'], model='additive', period=12)
# result.plot()
# plt.show()
trend = result.trend
seasonal = result.seasonal
residual = result.resid
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))

# Trend
plt.subplot(3,1,1)
plt.plot(trend.index, trend.values, color='blue')
plt.title('Trend Component')
plt.xlabel('Dates')
plt.ylabel('Trend')
plt.grid(True)

# Seasonal
plt.subplot(3, 1, 2)
plt.plot(seasonal.index, seasonal.values, color='orange')
plt.title('Seasonal Component')
plt.xlabel('Dates')
plt.ylabel('Seasonality')
plt.grid(True)

# Residual 
plt.subplot(3,1,3)
plt.plot(residual.index, residual.values, color='green')
plt.title('Residual Component')
plt.xlabel('Dates')
plt.ylabel('Residual')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
df_gas['Dates'] = pd.to_datetime(df_gas['Dates'])
df_gas.set_index('Dates', inplace=True)
```

```python
# filling out the past estimates of each day
df_gas = df_gas.resample('D').interpolate()
df_gas
```

```python
# forecasting (extrapolating one year into the future)
df_gas['TimeIndex'] = range(len(df_gas))
df_gas
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(df_gas[['TimeIndex']], df_gas['Prices'])

a = model.intercept_
b = model.coef_[0]

df_gas['Predicted'] = model.predict(df_gas[['TimeIndex']])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_gas.index, df_gas['Prices'], label='Actual', marker='o')
plt.plot(df_gas.index, df_gas['Predicted'], label='Trend Line', color='red')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
last_time_index = df_gas['TimeIndex'].iloc[-1]
future_time_index = np.arange(last_time_index + 1, last_time_index + 13)

future_prices = model.predict(future_time_index.reshape(-1, 1))
print(len(future_time_index)) 
```

```python
last_date = df_gas.index[-1]
future_dates = pd.date_range(start=last_date + pd.offsets.MonthEnd(1), periods=12, freq='M')
print(len(future_dates))        
```

```python
future_df = pd.DataFrame({'Date': future_dates, 'Predicted_Price': future_prices})
future_df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_gas.index, df_gas['Prices'], label='Actual')
plt.plot(df_gas.index, df_gas['Predicted'], label='Trend Line', color='red')
plt.plot(future_df.index, future_df['Predicted_Price'], label='Forecast (Next 12 months)', color='green', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# taking date input from the user
import pandas as pd

user_input = input("Enter date (YYYY-MM-DD): ")
try:
    input_date = pd.to_datetime(user_input)
    print("Parsed date:", input_date.date())
except Exception as e:
    print("Couldn't parse that date. Try formats like YYYY-MM-DD")  
```

```python
# checking the date
start_date = df_gas.index.min()
end_date = df_gas.index.max()
```

```python
if input_date < start_date:
    print('That date is before your available date, would need backward extrapolation')
elif input_date > end_date:
    print('That date is beyond your data, will forecast using the model')
else:
    print('That date is within your dataset, can estimate using itnerpolation')
    estimated_price = df_gas.loc[input_date, 'Prices'] if input_date in df_gas.index else \
                      df_gas['Prices'].interpolate().reindex(df_gas.index.union([input_date])).sort_index().loc[input_date]
    
    print(f"Estimated price on {input_date.date()}: {estimated_price:.2f}")
```

```python
# generating the predicted value
time_index_value = (input_date.year - start_date.year) * 12 + (input_date.month - start_date.month)
```

```python
predicted_price = model.predict([[time_index_value]])[0]
print(f"Estimated natural gas price on {input_date.date()}: ${predicted_price:.2f}")
```

```python
# checking how good the model is
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2 = r2_score(df_gas['Prices'], df_gas['Predicted'])
mse = mean_squared_error(df_gas['Prices'], df_gas['Predicted'])

print(f"R2 Score: {r2:.3f}")
print(f"Mean Squared Error: {mse:.3f}")
```

```python
# using both trend and seasonality of the data
df_gas['Month'] = df_gas.index.month
```

```python
# creating sine and cosine seasonal encodings
import numpy as np
df_gas['month_sin'] = np.sin(2 * np.pi * df_gas['Month'] / 12)
df_gas['month_cos'] = np.cos(2 * np.pi * df_gas['Month'] / 12)
```

```python
# defining x features and y target
X = df_gas[['TimeIndex', 'month_sin', 'month_cos']]
y = df_gas['Prices']
```

```python
# fitting the linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# predict using the smae features to check fit
df_gas['Predicted'] = model.predict(X)
```

```python
# visualising 
import matplotlib.pyplot as plt
plt.plot(df_gas.index, df_gas['Prices'], label='Actual')
plt.plot(df_gas.index, df_gas['Predicted'], label='Predicted (Trend + Seasonality)', color='red')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

raw = input("Enter date (YYYY-MM-DD): ")
try:
    input_date = pd.to_datetime(raw)
    print("Parsed date:", input_date.date())
except Exception as e:
    print("Couldn't parse that date. Try formats like YYYY-MM-DD")
```

```python
total_days = (df_gas.index[-1] - df_gas.index[0]).days
input_time_index = (input_date - df_gas.index[0]).days / total_days

month = input_date.month
month_sin = np.sin(2 * np.pi * month / 12)
month_cos = np.cos(2 * np.pi * month / 12)
```

```python
# predicting
X_new = [[input_time_index, month_sin, month_cos]]
predicted_price = model.predict(X_new)[0]
print(f"Estimated price for {input_date.date()}: ${predicted_price:.2f}")
```

```python
# checking the model again
from sklearn.metrics import r2_score, mean_squared_error
```
